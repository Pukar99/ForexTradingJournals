```python
import pandas as pd

# Read the CSV file
df = pd.read_csv("trading_journal_demo_data.csv")

# Extract the information inside parentheses from the 'Strategy' column
df['Strategy'] = df['Strategy'].str.extract(r'\((.*?)\)')

# Display the result
print(df)
```

```python
 #Extract profit and loss from the 'Profit/Loss' column
df['Profit/Loss'] = df['Profit/Loss'].str.extract(r'\$([\d,]+)').replace(',', '', regex=True).astype(float)
```

```python
# Calculate Total Profit, Total Loss, and Net Profit
total_profit = df.loc[df['Profit/Loss'] > 0, 'Profit/Loss'].sum()
total_loss = df.loc[df['Profit/Loss'] < 0, 'Profit/Loss'].sum()
net_profit = total_profit - total_loss
```

```python

# Find Highest Profit, Highest Loss, Best Strategy, Worst Strategy, Highest Lot
highest_profit = df.loc[df['Profit/Loss'].idxmax()]
highest_loss = df.loc[df['Profit/Loss'].idxmin()]
best_strategy = df.loc[df['Profit/Loss'].idxmax(), 'Strategy']
worst_strategy = df.loc[df['Profit/Loss'].idxmin(), 'Strategy']

# Convert 'Position Size' to numeric values
df['Position Size'] = pd.to_numeric(df['Position Size'], errors='coerce')

# Find the index with the maximum 'Position Size'
highest_lot_idx = df['Position Size'].idxmax()
highest_lot = df.loc[highest_lot_idx]

# Display Total Profit and Loss
print(f"Total Profit: ${total_profit:.2f}")
print(f"Total Loss: ${total_loss:.2f}")
print(f"Net Profit: ${net_profit:.2f}")

# Provide Insights
if net_profit > 0:
    print("Overall positive performance. Good job!")
elif net_profit < 0:
    print("Overall negative performance. Evaluate your strategies.")
else:
    print("Break-even. Consider refining your trading approach.")

# Additional insights
print(f"Highest Profit Trade:\n{highest_profit}")
print(f"Highest Loss Trade:\n{highest_loss}")
print(f"Best Strategy: {best_strategy}")
print(f"Worst Strategy: {worst_strategy}")
print(f"Highest Lot Size Trade:\n{highest_lot}")
```
